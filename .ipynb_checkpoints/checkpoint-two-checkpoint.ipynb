```python


plt.style.use('seaborn')
plt.figure(figsize=(20,7),facecolor='w')
 
plt.subplot(1,3,1)
plt.hist(mv_df['Total Library Size'],edgecolor='black',color='pink')
plt.xlabel("Size of the Library")
plt.ylabel("Distribution")
plt.title("Histogram for Library Size")

plt.subplot(1,3,2)
plt.hist(mv_df['No. of TV Shows'],edgecolor='black',color="lightgreen")
plt.xlabel("No. of TV Shows")
plt.ylabel("Distribution")
plt.title("Histogram for No. of TV Shows")
 
plt.subplot(1,3,3)
plt.hist(mv_df['No. of Movies'],edgecolor='black',color="cyan")
plt.xlabel("No. of Movies")
plt.ylabel("Distribution")
plt.title("Histogram for No. of Movies")

 
 
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(20,7),facecolor='w')
plt.subplot(1,3,1)
plt.hist(mv_df['Cost Per Month - Basic ($)'],edgecolor='black',color="cyan")
plt.xlabel("Cost Per Month - Basic ($)")
plt.ylabel("Distribution")
plt.title("Histogram for Cost Per Month - Basic ($")

plt.subplot(1,3,2)
plt.hist(mv_df['Cost Per Month - Standard ($)'],edgecolor='black',color='pink')
plt.xlabel("Cost Per Month - Standard ($)")
plt.ylabel("Distribution")
plt.title("Histogram for Cost Per Month - Standard ($)")


plt.subplot(1,3,3)
plt.hist(mv_df['Cost Per Month - Premium ($)'],edgecolor='black',color="lightgreen")
plt.xlabel("Cost Per Month - Premium ($)")
plt.ylabel("Distribution")
plt.title("Histogram forCost Per Month - Premium ($)")
```

```python
# For Year 2021
plt.style.use('seaborn')
plt.figure(figsize=(20,7),facecolor='w')
plt.subplot(1,3,1)
plt.hist(mv_df1['Cost Per Month - Basic ($)'],edgecolor='black',color="cyan")
plt.xlabel("Cost Per Month - Basic ($)")
plt.ylabel("Distribution")
plt.title("Histogram for Cost Per Month - Basic ($")

plt.subplot(1,3,2)
plt.hist(mv_df1['Cost Per Month - Standard ($)'],edgecolor='black',color='pink')
plt.xlabel("Cost Per Month - Standard ($)")
plt.ylabel("Distribution")
plt.title("Histogram for Cost Per Month - Standard ($)")


plt.subplot(1,3,3)
plt.hist(mv_df1['Cost Per Month - Premium ($)'],edgecolor='black',color="lightgreen")
plt.xlabel("Cost Per Month - Premium ($)")
plt.ylabel("Distribution")
plt.title("Histogram forCost Per Month - Premium ($)")
```

```python
plt.figure(figsize=(20,10),facecolor='w')
plt.plot(mv_df["Country"],mv_df["Cost Per Month - Basic ($)"],color="maroon",label="Basic Subscription")
plt.plot(mv_df["Country"],mv_df["Cost Per Month - Standard ($)"],color="darkblue",label="Standard Subscription")
plt.plot(mv_df["Country"],mv_df["Cost Per Month - Premium ($)"],color="orchid",label="Premium Subscription")
plt.xticks(rotation=90)
plt.title("All Subscription Costs in Various Countries",size=14)
plt.legend(title = "Subscription Type")
plt.show()
```

```python
#For 2021 Year Subscription Cost

plt.figure(figsize=(20,10),facecolor='w')
plt.plot(mv_df1["Country"],mv_df1["Cost Per Month - Basic ($)"],color="maroon",label="Basic Subscription")
plt.plot(mv_df1["Country"],mv_df1["Cost Per Month - Standard ($)"],color="darkblue",label="Standard Subscription")
plt.plot(mv_df1["Country"],mv_df1["Cost Per Month - Premium ($)"],color="orchid",label="Premium Subscription")
plt.xticks(rotation=90)
plt.title("All Subscription Costs in Various Countries 2021",size=14)
plt.legend(title = "Subscription Type 2021")
plt.show()
```
