```python
#For 2021 Year Subscription Cost

plt.figure(figsize=(20,10),facecolor='w')
plt.plot(mv_df1["Country"],mv_df1["Cost Per Month - Basic ($)"],color="maroon",label="Basic Subscription")
plt.plot(mv_df1["Country"],mv_df1["Cost Per Month - Standard ($)"],color="darkblue",label="Standard Subscription")
plt.plot(mv_df1["Country"],mv_df1["Cost Per Month - Premium ($)"],color="orchid",label="Premium Subscription")
plt.xticks(rotation=90)
plt.title("All Subscription Costs($) in Various Countries 2021",size=14)
plt.legend(title = "Subscription Type 2021")
plt.show()
```

```python
#to show values on Bars in seaborn

def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
#seaborn.barplot(*, x=None, y=None, hue=None, data=None, order=None, 
#hue_order=None, estimator=<function mean at 0x7ff320f315e0>, ci=95, n_boot=1000, units=None, 
#seed=None, orient=None, color=None, palette=None, saturation=0.75, errcolor='.26', errwidth=None, 
#capsize=None, dodge=True, ax=None, **kwargs)

plt.figure(figsize =(15,10))
ax=sns.barplot(x = mv_df1['Country'],y = mv_df1['Cost Per Month - Basic ($)'] ,data=mv_df1,palette ="summer")
plt.xlabel("Countries")
plt.ylabel("Basic Subscription Fee($)")
plt.xticks(rotation = 90)
plt.title("Cost of Basic Subscription For Diffrent Countries")
show_values(ax)
```
